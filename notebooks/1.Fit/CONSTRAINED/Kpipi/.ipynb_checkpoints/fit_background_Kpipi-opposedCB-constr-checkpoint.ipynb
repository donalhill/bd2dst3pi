# Name of the data: `data_KPiPi_constr`
# Fit to $m(D^*3\pi|m(D^*)=m(D^*)_{PDG})$ in the RapidSim $B^0 \to D^*K \pi^+\pi^-$ sample
- **Model**: sum of
    - A crystal ball with left powertail
    - A crystal ball with right powertail
- **Data** : `RapidSim` simulated of $B^0 \to D^*K \pi^+\pi^-$ with $K$ misidentified as a $\pi$ of same charge

```python
data_name = 'data_KPiPi_constr'
branch = 'B0_M_Tau_Pi12pip'
branch_constr = 'Dst_constr_B0_M_Tau_Pi12pip'
low = 4300
high = 5350

pos_text = 'RapidSim simulation'
```

## Libraries

```python
import zfit
import pandas as pd


from bd2dst3pi.locations import loc
from bd2dst3pi.definition import years, magnets
from bd2dst3pi.definition import latex_params_KPiPi as latex_params
from bd2dst3pi.pandas_root import load_data, add_constr_Dst

import sys
sys.path.append(loc.ROOT + 'library/')

from HEA.plot import plot_hist, plot_hist_auto
from HEA.plot.fit import plot_hist_fit_auto
from HEA.fit import json_to_latex_table, retrieve_params, show_latex_table
from HEA.fit.fit import launch_fit, save_params
from HEA.tools import show_dictionnary
```

## Load data

```python
branches = ['B0_M_Tau_Pi12pip', 'Dst_M', 'D0_M']
df = load_data(years, magnets, type_data='data_KPiPi', vars=branches)
for variable in branches: # GeV --> MeV
    df[variable] = 1000*df[variable]
```

```python
df = add_constr_Dst(df, branch)
```

### $\Delta M$ is already cut: $0.143\ GeV<\Delta M<0.148\ GeV$

```python
df['Delta_M'] =df['Dst_M']-df['D0_M']
print(len(df['Delta_M']))
print(len(df.query('Delta_M>143 and Delta_M<148')))

plot_hist({data_name:df}, 'Delta_M','$\Delta M$', 'GeV/$c^2$', n_bins = 100, folder_name = data_name, colors='black', bar_mode=True,
           pos_text_LHC={'ha':'right', 'text':pos_text})
```

```python
plot_hist_auto({data_name:df}, branch, n_bins = 100, folder_name = data_name, colors='black', bar_mode=False,
                      low=low, high=high, pos_text_LHC={'x':0.1, 'y':0.90, 'ha':'left', 'text':pos_text})
```

```python
plot_hist_auto({data_name:df}, branch_constr, n_bins = 100, folder_name = data_name, colors='black', 
                     bar_mode=False, low=low, high=high, pos_text_LHC={'x':0.1, 'y':0.90, 'ha':'left', 'text':pos_text})
```

## Fit to $m(D^*3\pi|m(D^*)=m(D^*)_{PDG})$

```python
obs = zfit.Space("x", limits=(low, high))
```

### Starting values of the parameters

```python
# fraction   0.8875     +/-   0.014       False
# alphaL     0.2414     +/-   0.014       False
# muL          5215     +/-    0.91       False
# nL          4.397     +/-    0.24       False
# sigmaL       19.2     +/-    0.61       False
# alphaR     -3.386     +/-    0.23       False
# muR          5183     +/-       3       False
# nR            0.1     +/-  0.0042       False
# sigmaR      39.41     +/-     1.5       False

muL = zfit.Parameter("muL", 5215, 5210, 5230)
muR = zfit.Parameter("muR", 5180, 5160, 5200)

sigmaL = zfit.Parameter("sigmaL", 20, 10., 50.)
sigmaR = zfit.Parameter("sigmaR", 40, 10., 50.)

alphaL = zfit.Parameter("alphaL", 0.3, 0.01, 2.)
alphaR = zfit.Parameter("alphaR", -1.8, -5., -0.1)

nL = zfit.Parameter("nL", 4.4, 0.1, 10.)
nR = zfit.Parameter("nR", 45., 0.000000000000000000000000000000000000000001, 60.)

frac = zfit.Parameter("fraction", 0.9, 0.1, 1.) # fraction between L and R
```

### PDFs

```python
# doubleCB = zfit.pdf.DoubleCB(mu, sigma, alphaL, alphaR, nL, nR, obs=obs)

crystalBallL = zfit.pdf.CrystalBall(muL, sigmaL, alphaL, nL, obs=obs)
crystalBallR = zfit.pdf.CrystalBall(muR, sigmaR, alphaR, nR, obs=obs)
# gaussianR = zfit.pdf.Gauss(muR, sigmaR, obs=obs)

model = zfit.pdf.SumPDF([crystalBallL, crystalBallR], fracs=frac) # crystalBallR is multiplied by fracs

data = zfit.Data.from_pandas(df[branch_constr],obs=obs)
```

### Fit

```python
result, params = launch_fit(model, data)
result, params = launch_fit(model, data) 
result, params = launch_fit(model, data) 
```

### Result of the fit

```python
plot_hist_fit_auto(df, branch_constr, models = [model, crystalBallL, crystalBallR], 
                       models_names=['','',''],
                       obs=obs, n_bins = 100,bar_mode = False,
                       folder_name = data_name+'_details', models_types= 'mss', loc_res='lower left',
                       data_name = data_name+'_params', params=params, latex_params=latex_params, show_chi2=True)   
```

```python
fig, ax, _ = plot_hist_fit_auto(df, branch_constr, models = [model, crystalBallL, crystalBallR], 
                                    models_names=['','',''], colors=['saddlebrown', 'orange', 'salmon'],
                                    obs=obs, n_bins = 70,bar_mode = False,
                                    folder_name = data_name, models_types= 'mss', loc_res='lower left',
                                    data_name = data_name, 
                                    pos_text_LHC={'x':0.05, 'y':0.70, 'ha':'left', 'text':pos_text})   
```
