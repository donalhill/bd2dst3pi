# MC vs background variables

Goal: 
- Compare the training variables for the MC (signal) and for the background samples (for the BDT training)
- Plot the correlation of the training variables (for the training of the BDTs)

```python
BDT_name = 'gradient'
low_bkg = 5400
high_bkg = 5570

branch_constr = 'Dst_constr_B0_M'
```

```python
import numpy as np
import matplotlib.pyplot as plt

from bd2dst3pi.locations import loc
from bd2dst3pi.definition import years, magnets
from bd2dst3pi.pandas_root import load_data, add_constr_Dst

import sys
sys.path.append(loc.ROOT + 'library/')

from HEA.plot import plot_hist_auto, plot_hist, plot_scatter2d_auto
from HEA.tools import show_dictionnary
from HEA.plot.tools import save_fig
from HEA.pandas_root import save_root, load_saved_root, get_dataframe_from_raw_branches_functions
import HEA.BDT as bt


from HEA import (
    get_branches_from_raw_branches_functions,
    get_raw_branches_from_raw_branches_functions
)
```

## Load data

### Training variables

```python
raw_branches_functions = [
    ('B0_DIRA_OWNPV', 'ln(1-x)'),
    (('B0_ENDVERTEX_CHI2', 'B0_ENDVERTEX_NDOF'), 'x/y'),
    'B0_PT',
    ('B0_FDCHI2_OWNPV', 'ln(x)'),
    'B0_IP_OWNPV', # much better with it
#     'B0_FD_OWNPV',

    ('Dst_DIRA_OWNPV', 'ln(1-x)'), # better with it
    'Dst_PT',
#     'Dst_FDCHI2_OWNPV', # worse with it
    'Dst_IPCHI2_OWNPV',
    'Dst_IP_OWNPV',
#     'Dst_FD_OWNPV', # different for sWeighted and MC

#     ('D0_DIRA_OWNPV', 'ln(1-x)'), # worst with it # different for sWeighted and MC
#     (('D0_ENDVERTEX_CHI2', 'D0_ENDVERTEX_NDOF'), 'x/y'),
    'D0_PT',
    ('D0_FDCHI2_OWNPV', 'ln(x)'),
    ('D0_IPCHI2_OWNPV', 'ln(x)'),
    'D0_IP_OWNPV',
#     'D0_FD_OWNPV', # different for sWeighted and MC

#     ('tau_DIRA_OWNPV', 'ln(1-x)'), # different for sWeighted and MC
    (('tau_ENDVERTEX_CHI2', 'tau_ENDVERTEX_NDOF'), 'x/y'), # a bit worse without it 
    'tau_PT',
    ('tau_FDCHI2_OWNPV', 'ln(x)'),
    ('tau_IPCHI2_OWNPV', 'ln(x)'),
    'tau_IP_OWNPV',
#     'tau_FD_OWNPV', # better without it # different for sWeighted and MC

    'Dst_pion_PT',
    ('Dst_pion_IPCHI2_OWNPV', 'ln(x)'),

    'D0_kaon_PT',
    ('D0_kaon_IPCHI2_OWNPV', 'ln(x)'),

    'D0_pion_PT', # better with it
    ('D0_pion_IPCHI2_OWNPV', 'ln(x)'),
       
#     'B0_ETA', # different for sWeighted and MC
#     'Dst_ETA', # different for sWeighted and MC
#     'D0_ETA', # different for sWeighted and MC
#     'tau_ETA', # different for sWeighted and MC
#     'Dst_pion_ETA', # different for sWeighted and MC
#     'D0_kaon_ETA', # different for sWeighted and MC
#     'D0_pion_ETA', # different for sWeighted and MC
]

raw_branches = get_raw_branches_from_raw_branches_functions(raw_branches_functions)
branches = get_branches_from_raw_branches_functions(raw_branches_functions)

    
add_branches = ['B0_M', 'tau_M', 'Dst_M', 'D0_M']
```

### Load dataframes

```python
df = {}
df['MC']     = load_data(years,magnets,type_data = 'MC'      ,vars = raw_branches+add_branches, cut_DeltaM=True, cut_PIDK=None)
df['training_MC'] = get_dataframe_from_raw_branches_functions(df['MC'], raw_branches_functions)
```

```python
df['common'] = load_data(years,magnets,type_data = 'common'  ,vars = raw_branches+add_branches, cut_DeltaM=True, cut_PIDK=None)
df['common'] = add_constr_Dst(df['common'])
add_branches += [branch_constr]
```

### Dataframes with the training variables

```python
df['training_common'] = get_dataframe_from_raw_branches_functions(df['common'], raw_branches_functions + add_branches)
```

```python
df['bkg'] = df['training_common'].query(f'{branch_constr} > {low_bkg} and {branch_constr} <{high_bkg}')
```

```python
print(branches)
```

```python
for branch in add_branches:
    for data in 'training_MC', 'bkg':
        try:
            df['bkg'] = df['bkg'].drop(branch, 1)
        except KeyError:
            pass
    
```

## Compare MC and background data

### Range of the variables

```python
range_dic = {
    'B0_PT': [0,35000],
    'Dst_PT':[0,30000],
    'tau_PT':[0,18000],
    'Dst_pion_PT':[0,2300],
    'Dst_kaon_PT':[0,2300],
    'D0_pion_PT':[0,20000],
    'D0_kaon_PT':[0,20000],
    'D0_PT':[0,3000],
    'B0_ENDVERTEX_CHI2':[0,60],
    'tau_ENDVERTEX_CHI2':[None,None],
    'B0_OWNPV_CHI2':[None,None],    
    'B0_FDCHI2_OWNPV':[0, 13000], 
    'B0_DIRA_OWNPV':[0.99996, 1], 
    'Dst_IPCHI2_OWNPV':[0, 2000], 
    'D0_kaon_IPCHI2_OWNPV':[0, 20000], 
    'D0_kaon_IPCHI2_OWNPV':[0, 20000], 
    'D0_pion_IPCHI2_OWNPV':[0, 20000], 
    'pion_IPCHI2_OWNPV':[0, 2000], 
    'tau_FDCHI2_OWNPV':[0, 200000], 
    'tau_FD_OWNPV':[0, 500], 
    'tau_pion0_TRACK_CHI2NDOF':[None,None], 
    'tau_pion1_TRACK_CHI2NDOF':[None,None], 
    'tau_pion2_TRACK_CHI2NDOF':[None,None],
    'D0_ENDVERTEX_CHI2,D0_ENDVERTEX_NDOF:x/y':[0, 10],
    'B0_ENDVERTEX_CHI2,B0_ENDVERTEX_NDOF:x/y':[0, 10],
    'B0_IP_OWNPV':[0, 0.1], 
    'D0_IP_OWNPV':[0, 2], 
    'Dst_IP_OWNPV':[0, 2], 
    'tau_IP_OWNPV':[0, 4], 
    'B0_FD_OWNPV':[0, 100], 
    'Dst_FD_OWNPV':[0, 200], 
    'tau_FD_OWNPV':[0, 150], 
    'D0_FD_OWNPV':[0, 200],
    "Dst_FDCHI2_OWNPV":[-1000, 20000],
}
```

### Plots

```python
for sub_branch in branches:
    print(f"=====================  {sub_branch} =====================")
    
    if sub_branch in range_dic:
        low = range_dic[sub_branch][0]
        high = range_dic[sub_branch][1]
    else:
        low = None
        high = None
    
    fig, ax = plot_hist_auto({'MC': df['training_MC'], 'Background': df['bkg']}, sub_branch, alpha=0.5, 
                                   n_bins=40, bar_mode = True, colors=['g', 'y'], low=low, high=high, 
                                   pos_text_LHC=[0.95, 0.95, 'right'], loc_leg = 'upper left',
                                   folder_name=f'BDT/{BDT_name}/variables',
                                   show_ncounts=False, save_fig=True)
    
```

## EXAMPLE - 2D scatter plots
(Illustration of MVA)

### Variables

```python
var1 = 'B0_FDCHI2_OWNPV:ln(x)'
var2 = 'D0_IPCHI2_OWNPV:ln(x)'

variables = [var1, var2, 'BDT', 'Dst_constr_B0_M']
```

### `common` dataset
I need to replace `__` by `:` because:
- `root_pandas` cannot save columns with `:`
- Thus, I saved them with `__` instead but I need to replace that by `:` so that my scripts recognise the variable

```python
df['common'] = load_saved_root(f'common_{BDT_name}_full', folder_name='common')

columns = list(df['common'].columns)
for column in columns:
    if '__' in column:
        df['common'][column.replace('__', ':')] = df['common'][column].values
        df['common'] = df['common'].drop(column, 1)
        
df['bkg'] = df['common'].query(f'{branch_constr} > {low_bkg} and {branch_constr} <{high_bkg}')
```

```python
df['common']
```

### Plotted dataframes and colors

```python
dfs = {'Background':df['bkg'], 'Signal (MC)':df['training_MC'].sample(11016, random_state=50)}
colors = ['r', 'g']
```

```python
dfs_LHCb = {'LHCb data':df['common'].sample(len(df['MC']), random_state=50), 'Signal (MC)':df['training_MC']}
colors_LHCb = ['b', 'g']
```

### 2D scatter plots of the two variables  in the training signal and background sample

```python
fig, ax = plot_scatter2d_auto(dfs, [var1, var2], folder_name='signal_vs_background', marker='.', s=1,
                        low=[2, None], high=[12.5, None], fontsize_label=17,
                       colors = colors)
```

```python
# Fixing random state for reproducibility
np.random.seed(19680801)

# the random data
x = np.random.randn(1000)
y = np.random.randn(1000)

#.definition for the axes
left, width = 0.1, 0.65
bottom, height = 0.1, 0.65
spacing = 0.005


rect_scatter = [left, bottom, width, height]
rect_histx = [left, bottom + height + spacing, width, 0.2]
rect_histy = [left + width + spacing, bottom, 0.2, height]

# start with a rectangular Figure
fig = plt.figure(figsize=(8, 8))

ax_scatter = plt.axes(rect_scatter)
ax_scatter.tick_params(direction='in', top=True, right=True)
ax_histx = plt.axes(rect_histx)
ax_histx.tick_params(direction='in', labelbottom=False)
ax_histy = plt.axes(rect_histy)
ax_histy.tick_params(direction='in', labelleft=False)


scs = plot_scatter2d_auto(dfs, [var1, var2], folder_name='signal_vs_background', marker='.', s=1,
                        low=[2, None], high=[12.5, None], ax=ax_scatter, fontsize_label=14,
                       colors=colors, get_sc=True)

plot_hist_auto(dfs, var1, folder_name='signal_vs_background',
                        low=2, high=12.5,  ax=ax_histx, show_leg=False, fontsize_label=10, bar_mode=True,
                       colors = colors)
ax_histx.set_xlabel(None)
plot_hist_auto(dfs, var2, folder_name='signal_vs_background',
               low=None, high=None, 
               ax=ax_histy, show_leg=False, orientation='horizontal',
               bar_mode=True,
               fontsize_label=10,
               colors=colors)
ax_histy.set_ylabel(None)

ax_scatter.set_xlim(ax_histx.get_xlim())
ax_scatter.set_ylim(ax_histy.get_ylim())

xticks = ax_scatter.xaxis.get_major_ticks()
xticks[-2].label1.set_visible(False)

yticks = ax_scatter.yaxis.get_major_ticks()
yticks[-2].label1.set_visible(False)
yticks[-1].label1.set_visible(False)

ax_scatter.xaxis.labelpad = 10
ax_scatter.legend(fontsize=20, markerscale=10)

xmin, xmax = ax_histx.get_xlim()
ymin, ymax = ax_histy.get_ylim()
x = [3.5, 9.5]
y = [ymin, ymax]

ax_scatter.plot(x, y, color='k', linewidth=1., linestyle='--') 
ax_scatter.fill(x + [xmin, xmin], y + [ymax, ymin], 'r', alpha=0.1)


for sc in scs:
    sc.set_rasterized(True)
    sc.set_zorder(-20)

    
    
save_fig(fig, 'scatter_plot_signal_bkg', folder_name='training_BDT')
```
