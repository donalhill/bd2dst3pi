# Test `adaboost_without_P_with_cut_deltaM` for 80% of ws data

## Load data

```python
name_BDT = 'adaboost_0.8_without_P_cutDeltaM'
```

```python
variables = ['B0_M','BDT']
```

```python
df = load_data(years,magnets,type_data = 'common',vars = variables)
```

```python
print(df)
```

### BDT plot

```python
plot_hist({'$BDT':df},'BDT', 'BDT', '', n_bins = 50,mode_hist=False, colors='orangered', 
            name_file = 'BDT',name_folder = name_BDT)
```

### `BO_M`

```python
low = 5150.
high = 5550.

plot_hist_particle({'no cut':df, '$BDT>0$':df.query('BDT>0'), 
            '$BDT<0$':df.query('BDT<0')},
          'B0_M', n_bins = 100,low = low,high=high,mode_hist=True, 
                       colors=['b', 'g', 'r'],
            name_file = 'BDT_cut_on_B0_M',name_folder = name_BDT, density=False)
```

```python
low = 5150.
high = 5550.

plot_hist_particle({'no cut':df, '$BDT>0.2$':df.query('BDT>0')},
          'B0_M', n_bins = 100,low = low,high=high,mode_hist=True, 
                       colors=['b', 'g', 'r'],
            name_file = 'BDT_cut_on_B0_M',name_folder = name_BDT, density=False)
```

```python
low = 5050.
high = 5550.

plot_hist_particle({'$BDT>0.1$':df.query('BDT>0.1'),
             '$BDT>0$':df.query('BDT>0'),
             '$BDT>-0.1$':df.query('BDT>-0.1')},
          'B0_M', low = low, high=high, mode_hist=True,
            name_file = 'BDT_different_cut_on_B0_M',name_folder=name_BDT)
```
