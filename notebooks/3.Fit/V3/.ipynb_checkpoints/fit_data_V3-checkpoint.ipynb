### Fitting

```python
def fit_ga(low_tau_M, high_tau_M, cut_BDT, df_BDT, initial_values):
    
    obs_tau_M = zfit.Space("x", limits=(low_tau_M, high_tau_M))
    
    zparams = {}
    
    for var_ga in initial_values.keys():
        zparams[var_ga] = zfit.Parameter(f"{var_ga}|{cut_BDT}", initial_values[var_ga]['value'], 
                                         initial_values[var_ga]['low'], initial_values[var_ga]['high'])
    
    gaussian_ga = zfit.pdf.Gauss(zparams["mu_ga"],zparams["sigma_ga"],obs=obs_tau_M)
    exponential_ga = zfit.pdf.Exponential(zparams["lambda_ga"], obs=obs_tau_M)
    
    gaussian_ext_ga = gaussian_ga.create_extended(zparams["n_sig_ga"])
    exp_ext_ga = exponential_ga.create_extended(zparams["n_bkg_ga"])
    model_ga = zfit.pdf.SumPDF([gaussian_ext_ga, exp_ext_ga]) # gauss is folded by frac
    
    data_tau_M = zfit.Data.from_pandas(df_BDT,obs=obs_tau_M)
    result_g, params_g = launch_fit(model_ga, data_tau_M, extended=True)
    
    return result_g, params_g, obs_tau_M, model_ga, gaussian_ext_ga, exp_ext_ga
```

```python
name_params_ga ={
    f'mu_ga|{cut_BDT}':'$\mu$',
    f'sigma_ga|{cut_BDT}':'$\sigma$',
    f'n_bkg_ga|{cut_BDT}':'$n_{bkg}$',
    f'n_sig_ga|{cut_BDT}':'$n_{sig}$',
    f'lambda_ga|{cut_BDT}':'$\\lambda$',
}
```

```python
result_g, params_g, obs_tau_M, model_ga, gaussian_ext_ga, exp_ext_ga = fit_ga(low_tau_M, high_tau_M, 
                                                                              cut_BDT, df[k]['tau_M'],
                                                                             d_initial_values_ga[k])

```

```python
plot_hist_fit_particle(df[k],'tau_M',models = [model_ga,gaussian_ext_ga,exp_ext_ga], 
              obs=obs_tau_M, n_bins = 100,mode_hist = False, name_data = 'data strip', cut_BDT=cut_BDT,
              params=params_g, name_params=name_params_ga)
```

```python
save_params(params_g,f'tau_M_B_BDT_sup_{cut_BDT}', True)
```

### Fitting

```python
obs = zfit.Space("x", limits=(low, high))
initial_values = d_initial_values[k]
zparams = {}
for var in initial_values.keys():
    zparams[var] = zfit.Parameter(f"{var}|{cut_BDT}",
                                  initial_values[var]['value'], 
                                  initial_values[var]['low'], 
                                  initial_values[var]['high'],
                                  floating = initial_values[var]['floating'])
    #print(zparams[var])
```

```python
doubleCB = zfit.pdf.DoubleCB(zparams['mu'], zparams['sigma'],
                             zparams['alphaL'], zparams['alphaR'],
                             zparams['nL'], zparams['nR'], 
                             obs=obs)

exponential = zfit.pdf.Exponential(zparams['lambda'], obs=obs)
gaussian = zfit.pdf.Gauss(zparams['mu2'],zparams['sigma2'],obs=obs)

doubleCB_Ds = zfit.pdf.DoubleCB(zparams['mu'], zparams['sigma'], 
                                zparams['alphaL'], zparams['alphaR'], 
                                zparams['nL'], zparams['nR'], 
                                obs=obs)

# sum of the two PDFs
doubleCB_ext = doubleCB.create_extended(zparams['n_sig'])
exp_ext = exponential.create_extended(zparams['n_bkg'])
gaussian_ext = gaussian.create_extended(zparams['n_bkg2'])
doubleCB_Ds_ext = doubleCB_Ds.create_extended(zparams['n_bkg_Ds'])

model = zfit.pdf.SumPDF([doubleCB_ext, exp_ext,gaussian_ext,doubleCB_Ds_ext]) # gauss is folded by frac

data = zfit.Data.from_pandas(df[k]['B0_M'], obs=obs)
```
