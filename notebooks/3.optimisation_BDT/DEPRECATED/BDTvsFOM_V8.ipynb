## Import

```python
import numpy as np
import json

from bd2dst3pi.locations import loc

import sys
sys.path.append(loc.SCRIPTS)
from plot.line import plot_x_list_ys

from uncertainties import ufloat, unumpy, umath
```

## Load data

I forgot to compute the error on $n_B$

$$ n_B = I \times n_{bkg}$$
So
$$\Delta n_B = I \Delta n_{bkg}= \frac{n_B}{n_{bkg}} \Delta n_{bkg}$$

```python
name_data = "all_common_cutDeltaM_V8"
```

```python
l_BDT = np.arange(-0.45,0.25,0.05) # list of the cuts on the BDT which the data has been fit on

lrB_B0Kpipi = [] # background B0
lS = [] # Signal B0 -> Ds 3pi
lB_combi = [] # combinatorial background

ltot = [] # S + B

squareRootlSplusB = []

for i, BDT in enumerate(l_BDT):
    BDT = round(BDT,2)
    if BDT==0:
        BDT=0
    
    with open(f'{loc.JSON}/{name_data}_BDT{BDT}_params.json', 'r') as f:
        params = json.load(f)
    
    lrB_B0Kpipi.append(ufloat(params[f'r_bkgK|BDT{BDT}'], params[f'r_bkgK|BDT{BDT}_err'])) 

    if 'n_B_err' not in params:
        params['n_B_err'] = params['n_B']/params[f'n_bkg|BDT{BDT}']*params[f'n_bkg|BDT{BDT}_err']

    lB_combi.append(ufloat(params['n_B'], params['n_B_err'])) # combinatorial
    
    lS.append(ufloat(params[f'n_sig|BDT{BDT}'],params[f'n_sig|BDT{BDT}_err']))
    
    ltot.append(lS[i]+lB_combi[i])
    squareRootlSplusB.append(umath.sqrt(ltot[i]))
    
lS = np.array(lS)
lB_combi = np.array(lB_combi)
lrB_B0Kpipi = np.array(lrB_B0Kpipi)
```

## Define FoM

```python
signal_purity = lS/ltot
signal_significance = lS/squareRootlSplusB
FoM = signal_purity*signal_significance
```
