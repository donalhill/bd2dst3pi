```python
import matplotlib.pyplot as plt
from bd2dst3pi.locations import loc

import numpy as np
import pandas as pd

from bd2dst3pi.definitions import years, magnets
import os.path as op

import sys
sys.path.append(loc.SCRIPTS)
from load_save_data import load_data, add_flight_distance_tau
import plot.histogram as h
from plot.tool import save_file
```

# Load variables

Needed variables:
- B0_M and BDT
- B0_ENDVERTEX_X, B0_ENDVERTEX_Y, B0_ENDVERTEX_Z
- tau_ENDVERTEX_X, tau_ENDVERTEX_Y, tau_ENDVERTEX_Z
- tau_ENDVERTEX_XERR, tau_ENDVERTEX_YERR, tau_ENDVERTEX_ZERR

```python
variables = []

for particle in 'B0', 'Dst', 'D0','tau':
    variables.append(f"{particle}_M")

for particle in 'B0', 'tau':
    for axis in 'X','Y','Z':
        variables.append(f'{particle}_ENDVERTEX_{axis}')
        variables.append(f'{particle}_ENDVERTEX_{axis}ERR')

variables.append('BDT')
```

```python
df = load_data(years,magnets,type_data = 'common',vars = variables, cut_DeltaM=True, cut_PIDK=None)
```

```python
name_data = 'all_common_cutDeltaM'
```

# B0_M vs BDT

```python
name_BDT = 'adaboost_0.8_without_P_cutDeltaM'
low_B0_M = 5150.
high_B0_M = 5550.
```

```python
fig, ax = h.plot_hist2d_particle(df,['B0_M','BDT'], low=[low_B0_M,None],high=[high_B0_M,None],
            name_folder = name_BDT, log_scale = True, save_fig=False)
ax.axvline(x=5279.53, color='green', linewidth=5.)
save_file(fig, "BDTvsB0_M",name_BDT)
```

```python
h.plot_hist({'data':df},'B0_M', 'B0\_M', 'MeV/$c^2$', n_bins = 50,mode_hist=True,  low = low_B0_M, high = high_B0_M,
            name_file = 'B0_M',name_folder = name_BDT, low = 5150.)
```
