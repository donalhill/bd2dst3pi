# Plot some interesting histograms
- $\Delta M = m(D^*)-m(D^0)$
- `B0_M`of the `MC`, `common` and `ws` data

```python
from bd2dst3pi.locations import loc
from bd2dst3pi.definition import years, magnets
from bd2dst3pi.pandas_root import load_data

import sys
sys.path.append(loc.ROOT + '/library/')
from HEA.plot import plot_hist_auto, plot_hist
import HEA.plot.tools as pt
```

## Load

```python
variables = []

for particle in 'B0', 'Dst', 'D0','tau':
    variables.append(f"{particle}_M")
```

```python
df    = load_data(years,magnets,type_data = 'common'  ,vars = variables, cut_DeltaM=False, cut_PIDK=None)
df_MC = load_data(years,magnets,type_data = 'MC'      ,vars = variables, cut_DeltaM=False, cut_PIDK=None)
#df_ws = load_data(years,magnets,type_data = 'ws_strip' ,vars = variables, cut_DeltaM=False, cut_PIDK=None)
```

```python
df["Delta_M"] = df["Dst_M"] - df["D0_M"]
df_MC["Delta_M"] = df_MC["Dst_M"] - df_MC["D0_M"]
#df_ws["Delta_M"] = df_ws["Dst_M"] - df_ws["D0_M"]
```

## Plot $\Delta M = m(D^*)-m(D^0)$

```python
plot_hist({'MC':df_MC, 'LHCb data':df},'Delta_M', '$\Delta M$', 'MeV/$c^2$',
              n_bins = 100, bar_mode = True, low=139.5, colors=["green", "blue"], 
            pos_text_LHC={'x':0.90, 'y':0.70, 'ha':'right', 
                          'text': 'LHCb simulation \n \n LHCb preliminary \n 2 fb$^{-1}$' }, 
            show_ncounts=False)
```
