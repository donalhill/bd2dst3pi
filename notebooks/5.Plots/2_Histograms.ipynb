```python
from bd2dst3pi.locations import loc
from bd2dst3pi.definitions import years, magnets


import sys
sys.path.append(loc.SCRIPTS)
from load_save_data import load_data
import plot.histogram as h

#Gives us nice LaTeX fonts in the plots
import matplotlib as mpl
from matplotlib import rc
#rc('font',**{'family':'serif','serif':['Roman']})
rc('text', usetex=True)
mpl.rcParams['axes.unicode_minus'] = False
```

## Load

```python
variables = []

for particle in 'B0', 'Dst', 'D0','tau':
    variables.append(f"{particle}_M")
```

```python
df    = load_data(years,magnets,type_data = 'common'  ,vars = variables, cut_DeltaM=False, cut_PIDK=None)
df_MC = load_data(years,magnets,type_data = 'MC'      ,vars = variables, cut_DeltaM=False, cut_PIDK=None)
#df_ws = load_data(years,magnets,type_data = 'ws_strip' ,vars = variables, cut_DeltaM=False, cut_PIDK=None)
```

```python
df["Delta_M"] = df["Dst_M"] - df["D0_M"]
df_MC["Delta_M"] = df_MC["Dst_M"] - df_MC["D0_M"]
#df_ws["Delta_M"] = df_ws["Dst_M"] - df_ws["D0_M"]
```

## Plot histograms

```python
h.plot_hist({'MC':df_MC, 'LHCb data':df},'Delta_M', '$\Delta M$', 'MeV/$c^2$',
              n_bins = 100, mode_hist = True, low=139.5, colors=["green", "blue"], pos_text_LHC=[0.90, 0.70,'right'], 
            label_ncounts=False)
```
