# Constrained $m(D^*)$

The goal of the document is to investigate the use of constrained $m(D^*)$.
We'll look at $m(B^0)-m(D^*)+m(D^*)_{PDG}$. 

A more thorough analysis would define the constrained $m(B^0)$ by fixing $m(D*)$ at its PDG measured value, and redoing the reconstruction from this constrain.

Plotted:
- $m(B^0)-m(D^*)+m(D^*)_{PDG}$ of the `MC` data
- $m(B^0)-m(D^*)+m(D^*)_{PDG}$ of the `common` data

## Import

```python
import pandas as pd 
from bd2dst3pi.locations import loc
from bd2dst3pi.definition import years, magnets
from bd2dst3pi.pandas_root import load_data

import sys
sys.path.append(loc.ROOT + '/library/')
from HEA.plot import plot_hist_auto, plot_hist
import HEA.plot.tools as pt
```

## Load

```python
branches = []

for particle in 'B0', 'Dst', 'D0','tau':
    branches.append(f"{particle}_M")
```

```python
df = {}
df['common'] = load_data(years,magnets,type_data = 'common'  ,vars=branches, cut_DeltaM=True, cut_PIDK=None)
df['MC']     = load_data(years,magnets,type_data = 'MC'      ,vars=branches, cut_DeltaM=True, cut_PIDK=None)
```

```python
df['common_noDeltaM'] = load_data(years, magnets, type_data='common'  , vars= branches, cut_DeltaM=False, cut_PIDK=None)
```

[Charmed, strange mesons](https://pdg.lbl.gov/2020/tables/rpp2020-tab-mesons-charm.pdf), P.A. Zyla et al. (Particle Data Group), Prog. Theor. Exp. Phys. 2020, 083C01 (2020)

$m(D_s^{*\pm}(2010)) = 2010.26 \pm 0.05\ MeV$

```python
Dst_M_PDG = 2010.26
```

## Invariant mass Plot

### MC data

#### $m(D^*)$

```python
fig, ax = plot_hist_auto({'MC':df['MC']}, 'Dst_M',
              n_bins=100, bar_mode=False, colors='black', save_fig=False)

ax.axvline(x=Dst_M_PDG, color='red', linewidth=1.)
pt.save_fig(fig, 'Dst_M_MC')
```

```python
fig, ax = plot_hist_auto({'common':df['common']}, 'Dst_M', pos_text_LHC='right',
              n_bins = 100, bar_mode = False, colors='black', save_fig=False)

ax.axvline(x=Dst_M_PDG, color='red', linewidth=1.)
pt.save_fig(fig, 'Dst_M_common')
```

```python
fig, ax = plot_hist_auto({'common_noDeltaM':df['common_noDeltaM']}, 'Dst_M', pos_text_LHC='right',
              n_bins = 100, bar_mode = False, colors='black', save_fig=False)

ax.axvline(x=Dst_M_PDG, color='red', linewidth=1.)
pt.save_fig(fig, 'Dst_M_common_noCutDeltaM')
```

## Constrained $m(B^0)$

I create a new dataset for the "constrained" $m(B^0)$ to adapt the code to my way of plotting histograms with `plot_hist`
- `df['common_constr']['B0_M']`
- `df['MC_constr']['B0_M']`

```python
for d in 'MC', 'common':
    data_name = d + '_constr'
    df[data_name] = pd.DataFrame()
    df[data_name]["B0_M"] = df[d]["B0_M"] - df[d]['Dst_M'] + Dst_M_PDG
```

## Plot histograms

### MC

```python
low_MC = 5150.
high_MC = 5400.
```

```python
plot_hist_auto({f'MC with $m(D^*)$ = {Dst_M_PDG} MeV': df['MC_constr'],'MC':df['MC']}, 'B0_M',
              n_bins = 80, bar_mode = True, low=low_MC, high=high_MC, colors=['g', 'b'], density=False, 
                     show_ncounts=False)
```
