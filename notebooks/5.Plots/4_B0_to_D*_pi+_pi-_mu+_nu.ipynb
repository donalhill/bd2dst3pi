# Check $B^0 \to D^{*-} \pi^+ \pi^- \mu^+ \nu$


Plot the $m(D^*3\pi|m(D^*)=m(D^*)_{PDG})$ cutting on the probability that one of the pion having a sign opposite to the $D^*$ (i.e., `tau_pion0` or `tau_pion2`) is a muon.

We might be able to see the contribution of $B^0 \to D^{*-} \pi^+ \pi^- \mu^+ \nu$

```python
var = 'Dst_constr_B0_M'
low = 5150.
high = 5545.
```

## Import

```python
from bd2dst3pi.locations import loc
from bd2dst3pi.definitions import years, magnets
import pandas as pd


import sys
sys.path.append(loc.SCRIPTS)
from load_save_data import load_data, add_constr_Dst
import plot.histogram as h
from plot.tool import save_file
```

## Load

```python
variables = []

for particle in 'B0', 'Dst':
    variables.append(f"{particle}_M")

variables.append('tau_pion0_ProbNNmu')
variables.append('tau_pion1_ProbNNmu')
variables.append('tau_pion2_ProbNNmu')
variables.append('tau_pion0_ID')
variables.append('tau_pion1_ID')
variables.append('tau_pion2_ID')
variables.append('Dst_ID')
```

```python
df = load_data(years,magnets,type_data = 'common'  ,vars=variables, cut_DeltaM=True)
```

```python
df = add_constr_Dst(df)
```

## ID variable

```python
print(all(df['tau_pion0_ID'] == -df['Dst_ID']/413*211))
print(all(df['tau_pion1_ID'] == df['Dst_ID']/413*211))
print(all(df['tau_pion2_ID'] == -df['Dst_ID']/413*211))
```

- `tau_pion1_ID`: same charge as $D^*$
- `tau_pion0_ID`: opposite charge to $D^*$
- `tau_pion2_ID`: opposite charge to $D^*$

If $D^{*-}$:
- `tau_pion1_ID`: $\pi^-$
- `tau_pion0_ID`: $\pi^+$
- `tau_pion2_ID`: $\pi^+$

### `ProbNNmu` plots

```python
fig, ax = h.plot_hist_particle({'common':df}, 'tau_pion0_ProbNNmu', pos_text_LHC='right',
              n_bins = 100, mode_hist = False, colors='black', save_fig=False)
```

```python
df['tau_pion0_ProbNNmu']
```

## Cut to get $B^0 \to D^{*-} \pi^+ \pi^- \mu^+ \nu$

```python
df_cut_bkg = df.query('tau_pion0_ProbNNmu > -400 or tau_pion2_ProbNNmu > -400')
df_cut_bkg_plus = df_cut_bkg.query('not(tau_pion0_ProbNNmu < -400 and tau_pion2_ProbNNmu < -400)')

```

```python
print(f"Efficiency of the cut: {1 - len(df_cut_bkg)/len(df)}")
print(f"Efficiency of the second cut: {1 - len(df_cut_bkg_plus)/len(df_cut_bkg)}")
```
