# Check $B^0 \to D^{*-} \pi^+ \pi^- \mu^+ \nu$


Plot the $m(D^*3\pi|m(D^*)=m(D^*)_{PDG})$ cutting on the probability that one of the pion having a sign opposite to the $D^*$ (i.e., `tau_pion0` or `tau_pion2`) is a muon.

We might be able to see the contribution of $B^0 \to D^{*-} \pi^+ \pi^- \mu^+ \nu$

```python
branch_constr = 'Dst_constr_B0_M'
low = 5150.
high = 5545.
```

## Import

```python
import pandas as pd 
from bd2dst3pi.locations import loc
from bd2dst3pi.definition import years, magnets
from bd2dst3pi.pandas_root import load_data, add_constr_Dst

import sys
sys.path.append(loc.ROOT + '/library/')
from HEA.plot import plot_hist_auto, plot_hist, plot_hist2d_auto
import HEA.plot.tools as pt
```

## Load

```python
branches = []

for particle in 'B0', 'Dst':
    branches.append(f"{particle}_M")

branches.append('tau_pion0_ProbNNmu')
branches.append('tau_pion1_ProbNNmu')
branches.append('tau_pion2_ProbNNmu')
branches.append('tau_pion0_ID')
branches.append('tau_pion1_ID')
branches.append('tau_pion2_ID')
branches.append('Dst_ID')
```

```python
df = load_data(years,magnets,type_data = 'common'  ,vars=branches, cut_DeltaM=True)
```

```python
df = add_constr_Dst(df)
```

## ID variable

```python
print(all(df['tau_pion0_ID'] == -df['Dst_ID']/413*211))
print(all(df['tau_pion1_ID'] == df['Dst_ID']/413*211))
print(all(df['tau_pion2_ID'] == -df['Dst_ID']/413*211))
```

- `tau_pion1_ID`: same charge as $D^*$
- `tau_pion0_ID`: opposite charge to $D^*$
- `tau_pion2_ID`: opposite charge to $D^*$

If $D^{*-}$:
- `tau_pion1_ID`: $\pi^-$
- `tau_pion0_ID`: $\pi^+$
- `tau_pion2_ID`: $\pi^+$

### `ProbNNmu` plots

```python
fig, ax = plot_hist_auto({'common':df}, 'tau_pion0_ProbNNmu', pos_text_LHC='right',
              n_bins = 100, bar_mode = False, colors='black', save_fig=False)
```

```python
df['tau_pion0_ProbNNmu']
```

## Cut to get $B^0 \to D^{*-} \pi^+ \pi^- \mu^+ \nu$

```python
df_cut_bkg = df.query('tau_pion0_ProbNNmu > -400 or tau_pion2_ProbNNmu > -400')
df_cut_bkg_plus = df_cut_bkg.query('not(tau_pion0_ProbNNmu < -400 and tau_pion2_ProbNNmu < -400)')
```

```python
print(f"Efficiency of the cut: {1 - len(df_cut_bkg)/len(df)}")
print(f"Efficiency of the second cut: {1 - len(df_cut_bkg_plus)/len(df_cut_bkg)}")
```

```python
all(df_cut_bkg['tau_pion2_ProbNNmu']!=df_cut_bkg['tau_pion0_ProbNNmu'])
```

```python
df_cut_bkg['tau_pion0_ProbNNmu']
```

```python
fig, ax = plot_hist2d_auto(df_cut_bkg, ['tau_pion0_ProbNNmu', 'tau_pion2_ProbNNmu'],
                                 save_fig=False)
```
